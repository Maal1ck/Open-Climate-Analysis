# Contexte


La classification de Köppen, proposée par le climatologue allemand Wladimir Köppen en **1900** (et modifiée par la suite), est l'une des méthodes les plus utilisées pour décrire les climats de la planète. Elle repose principalement sur les températures, les précipitations et la saisonnalité pour classer les régions climatiques. Cette classification est appréciée pour sa simplicité et sa corrélation avec la végétation naturelle.

Les 5 grands types climatiques de Köppen
##A : Climat tropical

Températures élevées toute l'année (moyenne annuelle > 18 °C).
Précipitations abondantes, mais leur répartition varie selon le sous-type.
Sous-types :
Af : Climat équatorial (pluie toute l'année).
Am : Climat de mousson (saison sèche courte).
Aw : Climat tropical à saison sèche (savane).
B : Climat sec

##Hivers doux (température du mois le plus froid comprise entre -3 °C et 18 °C).

Sous-types :

Sous-types :
ET : Climat de toundra (moyenne du mois le plus chaud entre 0 et 10 °C).
EF : Climat glacial (aucun mois > 0 °C).


Avantages de la classification de Köppen:
Elle est simple à utiliser et directement basée sur des données climatiques mesurables (températures, précipitations).
Elle permet d'établir un lien direct avec les types de végétation naturelle.

Limites:
Elle ne prend pas en compte certains facteurs climatiques importants, comme les vents ou l'humidité relative.
Elle ne reflète pas toujours les impacts locaux du relief ou des microclimats.

## **Introduction**
NASA's Prediction Of Worldwide Energy Resources (POWER) project provides publicly accessible environmental data to support various applications, including renewable energy, sustainable development, and climate research. This presentation will explore what NASA POWER is, its key features, data applications, and how users can access and utilize the data.

---

## **1. What is NASA POWER?**
NASA POWER (Prediction Of Worldwide Energy Resources) is a project under NASA’s Earth Science Division designed to provide global climatology data. It primarily serves three user communities:
- **Renewable Energy (RE)**: Solar and wind energy assessments.
- **Sustainable Buildings (SB)**: Climate data for energy efficiency and building design.
- **Agriculture (AG)**: Environmental parameters for crop modeling and water resource management.

POWER delivers solar, meteorological, and climatological data derived from NASA’s satellite observations, reanalysis models, and atmospheric simulations.

---

## **2. Key Features of NASA POWER Data**
### **a. Global Coverage**
- Data is available globally, from **90°N to 90°S** latitude.
- High-resolution gridded data (~1° x 1° resolution).

### **b. Long-Term Historical Data**
- Covers several decades, with data available from **1981 to the present**.
- Includes real-time updates and historical climatology records.

### **c. Wide Range of Climate and Energy Parameters**
- **Solar Radiation Data**: Direct Normal Irradiance (DNI), Global Horizontal Irradiance (GHI), etc.
- **Meteorological Data**: Temperature, humidity, wind speed, pressure.
- **Cloud and Atmospheric Data**: Precipitation, cloud cover, aerosols.

### **d. Free and Open Access**
- No cost to users.
- Available through web portals, APIs, and downloadable files.

---

## **3. Applications of NASA POWER Data**
### **a. Renewable Energy**
- **Solar Energy**: Assessing solar power potential for photovoltaic (PV) and concentrated solar power (CSP) systems.
- **Wind Energy**: Evaluating wind resource availability for turbine placement.

### **b. Sustainable Buildings and Climate Monitoring**
- **Energy Efficiency**: Supporting passive cooling and heating designs.
- **Urban Planning**: Understanding local climate conditions for infrastructure planning.

### **c. Agriculture and Ecological Research**
- **Crop Modeling**: Providing inputs for agricultural growth simulations.
- **Water Management**: Assessing evaporation and precipitation trends.

### **d. Climate Change Research**
- **Trend Analysis**: Investigating long-term climate patterns.
- **Extreme Weather Studies**: Monitoring droughts, storms, and heatwaves.

---

## **4. How to Access NASA POWER Data**
### **a. POWER Data Access Tools**
NASA provides multiple ways to retrieve POWER data:

#### **1. POWER Web Portal**
- **Website**: [POWER Data Access Viewer](https://power.larc.nasa.gov/)
- **Features**:

  - Select data parameters, location, and time range.

#### **2. NASA POWER API**

#### **3. GIS and Remote Sensing Platforms**
- POWER data is compatible with **Google Earth Engine (GEE), ArcGIS, and QGIS** for spatial analysis.

#### **4. Integration with Python & R**

## NASA POWER DATA API DATA RETRIEVING


```python
import requests
import pandas as pd

# Define API endpoint
base_url = "https://power.larc.nasa.gov/api/temporal/daily/point"

# Define parameters
params = {
    "parameters": "PRECTOT,T2M",  # Total Precipitation & Temperature at 2m
    "community": "RE",  # Renewable Energy Community
    "longitude": float(input("Donner la longitude de la zone d'etude: ")),  # Longitude
    "latitude": float(input("Donner la latitude de la zone d'etude: ")),  # Latitude
    "start": "19810101",  # Start Date (YYYYMMDD)
    "end": "20241231",  # End Date (YYYYMMDD)
    "format": "JSON",  # Response format
}

# Make the request
response = requests.get(base_url, params=params)

# Check if request was successful
if response.status_code == 200:
    data = response.json()["properties"]["parameter"]

    # Convert data to Pandas DataFrame
    data = pd.DataFrame(data)
    data.index = pd.to_datetime(data.index)  # Convert index to datetime

    # Rename columns
    data = data.rename(columns={"PRECTOT": "Precipitation (mm)", "T2M": "Temperature (°C)"})

    # print(df.head())  # Display first few rows
else:
    print("Error:", response.status_code, response.text)
data = data.reset_index()
data = data.rename(columns = {'index':'Date','PRECTOTCORR':'Precipitation','Temperature (°C)':'Temperature'})
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
display(data)
```

```python
Temperature = data[['Year','Month','Temperature']]
Temperature_monthly = Temperature.groupby(['Year','Month'])['Temperature'].mean().reset_index()
Temperature_monthly = Temperature_monthly.pivot_table(values = 'Temperature', index = 'Year', columns = 'Month')
import calendar
Temperature_monthly.columns = [calendar.month_name[i] if isinstance(i, int) else i for i in Temperature_monthly.columns]
display(Temperature_monthly)

Precipitation = data[['Year','Month','Precipitation']]
Precipitation_monthly = Precipitation.groupby(['Year','Month'])['Precipitation'].sum().reset_index()
Precipitation_monthly = Precipitation_monthly.pivot_table(values = 'Precipitation', index = 'Year', columns = 'Month' )
Precipitation_monthly.columns = [calendar.month_name[i] if isinstance(i, int) else i for i in Precipitation_monthly.columns]
display(Precipitation_monthly)
```

```python
4
def Anomalies_PREC_TEMP():

    #1 Temperature anomalies

    # We are taking the raw data for both precipitation and because we will need the whole time series (i.e 1970 to 2016)
    TEMP_TRANSPOSED = Temperature_monthly.T
    DATA = TEMP_TRANSPOSED.mean().round(2)
    TEMP_MEAN_YEAR = DATA.to_frame(name = 'MEAN').reset_index()
    # TEMP_MEAN_YEAR = Temperature_monthly.groupby('Year')['Temperature'].mean().reset_index()
    PREC_TRANSPOSED = Precipitation_monthly.T
    DATA = PREC_TRANSPOSED.sum().round(2)
    PREC_SUM_YEAR = DATA.to_frame(name = 'SUM').reset_index()
    # PREC_SUM_YEAR = Precipitation_monthly.groupby('Year')['Precipitation'].sum().reset_index()

    #Computing the gap  T = Temp - Normal
    TEMP_MEAN_YEAR['Ecart'] =  TEMP_MEAN_YEAR['MEAN'] - Tann


    #Calcul de l'ecart RR = Temp - Normal
    PREC_SUM_YEAR['Ecart'] = 100*((PREC_SUM_YEAR['SUM'] - Pann) / Pann)

    # display(TEMP_MEAN_YEAR)
    # display(PREC_SUM_YEAR)
    years = PREC_SUM_YEAR['Year']

    #PLOTING THE TEMPERATURE GAP
    # global fig, ax
    fig, ax = plt.subplots(figsize = (15,5))
    plt.style.use("dark_background")
    title = f"Evolution temporelle de l'écart de la température moyenne annuelle \n p/r à la Normale {Starting_year}-{Ending_year} : {Location}"


    ax.plot(TEMP_MEAN_YEAR['Year'],TEMP_MEAN_YEAR['Ecart'], color = 'blue', marker = 'o') #label = 'Ecart'
    ax.set_xlabel('Year')
    ax.set_ylabel('Anomalies in Degree Celsius')
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.xticks(years[:], rotation=90)
    ax.axhline(0, color='yellow', linestyle='--', linewidth=1, marker = 'o')
    ax.set_title(title)
    # plt.savefig(f"ECART_TEMP_moyenne annuelle de {Location} p/r {Starting_year}_{Ending_year}.png")




    #2  TEMPERATURE TENDANCIES
    years = TEMP_MEAN_YEAR['Year']
    temperature = TEMP_MEAN_YEAR['MEAN']
    slope, intercept, r_value, p_value, std_err = linregress(years, temperature)

    # Create regression line
    regression_line = slope * TEMP_MEAN_YEAR['Year'] + intercept
    plt.figure(figsize=(15, 5))
    plt.plot(years, temperature, label="Annual Temperature", color="white", linewidth=2)
    plt.plot(years, regression_line, label=f"Trend: y={slope:.4f}x + {intercept:.2f}", color="red", linewidth=2)
    plt.title("Temperature: {:}\nSlope = {:.5f} Intercept = {:.2f}\nAdj R² = {:.5f}".format(Location,
    slope, intercept, r_value**2), fontsize=12)
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Temperature moyenne annuelle (°C)", fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.legend(fontsize=10)
    plt.xticks(years[:], rotation=90)

    plt.tight_layout()
    plt.show()
    # plt.savefig(f"Temperature tendency {Location} .png")


    #3 PRECIPITATION TENDENCY
    #RETRIEVING THE GAPS(Ecart in PREC_SUM_YEAR) to isolate them and plot them
    Ecarts = PREC_SUM_YEAR.loc[:,'Ecart']
    # SEPERATING THE VALUES IN DIFFERENT LISTS
    Excedents = [v if v>0 else 0 for v in Ecarts]
    Deficits = [v if v<0 else 0 for v in Ecarts]
    fig, ax = plt.subplots(figsize = (15,5))

    title = f"Evolution temporelle de l'écart de la précipitation annuelle \n p/r à la Normale {Starting_year}-{Ending_year} : {Location}"
    X_axis = years
    Y_axis_excedent = Excedents
    Y_axis_deficit = Deficits
    ax.bar(X_axis,Y_axis_excedent, color = 'red', label = 'Excedents')
    ax.bar(X_axis,Y_axis_deficit, color = 'blue', label = 'Deficits')
    ax.set_xlabel('Year')
    ax.set_ylabel('Relative Gap in %')
    plt.xticks(years[:], rotation=90)
    ax.axhline(50, color='white', linestyle='--', linewidth=1)
    ax.axhline(-50, color='white', linestyle='--', linewidth=1)
    ax.set_title(title)
    plt.legend()
    # plt.savefig(f"ECART_PREC_CUMUL annuelle de {Location} p/r {Starting_year}_{Ending_year}.png")

```
